```python
import pandas as pd

df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Desktop\internship\messy_data.csv")
df.head()
```

```python
df.columns
```

```python
# (1) Find and display all missing values

print(df.isnull().sum())
```

```python
#I ran this first to identify exactly which columns have missing data (NaNs) 
# so I could decide whether to drop those rows or fill them with a specific value.
```

```python
# (2) Drop rows where a key column (CustomerName) is missing
# Why: A record without a CustomerName is often considered incomplete or invalid for sales reporting.
initial_rows = len(df)
df = df.dropna(subset=['CustomerName'])
print(f"Rows before: {initial_rows}, Rows after: {len(df)}")
print("\n")
```

```python
#I chose to drop these rows because 'CustomerName' is a critical identifier; 
# without it, the record is incomplete and useless for customer-based analysis.
```

```python
# (3) Fill missing values in 'Quantity' with the column mean
# Why: Since quantity is a numeric value, filling with the mean allows us to keep the row 
# without significantly skewing the total sales metrics.
mean_val = df['Quantity'].mean()
df['Quantity'] = df['Quantity'].fillna(mean_val)
print(f"Filled missing Quantity with mean: {mean_val:.2f}")
print("Missing values in Quantity now:", df['Quantity'].isnull().sum())
print("\n")
```

```python
#I used the column mean to fill missing quantities because it preserves the 
# row while maintaining the statistical integrity of the distribution for that column.
```

```python
# (4) Remove duplicate rows
# Why: Duplicate entries lead to double-counting of revenue and orders.
initial_rows = len(df)
df = df.drop_duplicates()
print(f"Rows before: {initial_rows}, Rows after: {len(df)}")
print("\n")
```

```python
#I removed duplicates to ensure each order is unique, which prevents 
# double-counting revenue and skewing our final sales metrics.
```

```python
#(5) Fix 'Price' column datatype
# Why: 'Price' must be numeric for calculations like total revenue.
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
# Fill remaining NaNs introduced by the conversion with the median
df['Price'] = df['Price'].fillna(df['Price'].median())
print("New dtype of Price:", df['Price'].dtype)
print("Sample of cleaned Price column:\n", df['Price'].head())
```

```python
#I forced this conversion because the column contained text ('abc') 
# preventing math operations. By coercing errors to NaN and filling them with 
# the median, I made the column usable for calculations without losing data.
```
