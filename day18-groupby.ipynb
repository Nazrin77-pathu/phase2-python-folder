```python
import pandas as pd

data = {
    "OrderID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Phone","Tablet","Shoes","Shoes","Watch","Phone","Laptop"],
    "Category": ["Electronics","Electronics","Electronics","Fashion","Fashion","Accessories","Electronics","Electronics"],
    "City": ["Delhi","Mumbai","Delhi","Bangalore","Bangalore","Chennai","Delhi","Mumbai"],
    "Sales": [55000,30000,20000,4000,4000,5000,30000,55000],
    "Quantity": [1,2,1,2,2,1,2,1]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Category")["Sales"].count()
```

```python
df["Category"].value_counts()
```

```python
df.groupby("Category")["Sales"].agg(["sum", "mean", "count"])
```

```python
## Reflection
I used groupby() to calculate total, average, and count of sales for each category.
The total sales values matched my Power BI bar chart.
This confirms that both Pandas and Power BI perform the same aggregation operations.
Power BI is easier for visualization, while Pandas provides more flexibility for data analysis.
```
