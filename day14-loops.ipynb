```python
Scores=[46,68,83,49,90,54,38,71,62,25]
pass_mark=50

pass_count=0
fail_count=0

for score in Scores:
    if score>=pass_mark:
        print(score,"-Pass")
        pass_count+=1
    else:
        print(score,"-Fail")
        fail_count+=1

print(pass_count,"Students passed out of 10")
print(fail_count,"Students failed out of 10")
```
