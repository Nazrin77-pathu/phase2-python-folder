```python
def celcius_to_fahrenheit(temp):
    return(temp*9/5)+32

print(celcius_to_fahrenheit(0))
print(celcius_to_fahrenheit(25))
print(celcius_to_fahrenheit(100))

def get_grade(score):
    if score>=90:
        return "A"
    elif score>=75:
        return "B"
    elif score>=50:
        return "C"
    else:
        return "F"

print(get_grade(95))
print(get_grade(80))
print(get_grade(65))
print(get_grade(50))
print(get_grade(30))

```

##explanation of functions
### 1.celsius_to_fahrenheit(temp)
This fn takes a temperature in celsius as input.
It converts the temperature into Fahrenheit using the formula:
F = (C × 9/5) + 32
It then returns the converted temperature.

### 2. get_grade(score)
This function takes a student's score as input.
It checks the score using conditions:
- If score is 90 or above, it returns "A"
- If score is 75 or above, it returns "B"
- If score is 50 or above, it returns "C"
- If score is below 50, it returns "F"
